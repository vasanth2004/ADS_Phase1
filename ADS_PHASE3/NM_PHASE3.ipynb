```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/cpcb_dly_aq_tamil_nadu-2014.csv')
df.head()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['SO2'], bins=20)
plt.title('Histogram of SO2')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['City/Town/Village/Area'], df['RSPM/PM10'], alpha=0.5, color='green')
plt.title('Scatter Plot ')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
df['Sampling Date'] = pd.to_datetime(df['Sampling Date'])
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x='Sampling Date', y='NO2', data=df)
plt.xlabel('Category')
plt.ylabel('Numeric Value')
plt.show()
```
