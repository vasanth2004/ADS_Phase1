```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('AirQualityData.csv')
df.head()
```

```python
#i)Air quality analysis :Calculate average SO2, NO2, and RSPM/PM10 levels across different monitoring stations, cities, or areas.
# Group the data by "Location of Monitoring Station"
grouped_data = df.groupby("Location of Monitoring Station")
```

```python
# Calculate the mean for each column
mean_so2 = grouped_data["SO2"].mean()
mean_no2 = grouped_data["NO2"].mean()
mean_rspm_pm10 = grouped_data["RSPM/PM10"].mean()
mean_pm25 = grouped_data["PM 2.5"].mean()
```

```python
# Display the results
print("Mean SO2 levels:")
print(mean_so2)

```

```python
print("Mean NO2 levels:")
print(mean_no2)

```

```python
print("Mean RSPM/PM10 levels:")
print(mean_rspm_pm10)

```

```python
#i) Identify pollution trends and areas with high pollution levels.
mean_data = grouped_data[["SO2", "NO2", "RSPM/PM10", "PM 2.5"]].mean()
# locations with the highest average pollution levels.

highest_so2 = mean_data["SO2"].idxmax()
highest_no2 = mean_data["NO2"].idxmax()
highest_rspm_pm10 = mean_data["RSPM/PM10"].idxmax()
highest_pm25 = mean_data["PM 2.5"].idxmax()

#actual mean values for the highest pollution locations
mean_so2_highest = mean_data.loc[highest_so2, "SO2"]
mean_no2_highest = mean_data.loc[highest_no2, "NO2"]
mean_rspm_pm10_highest = mean_data.loc[highest_rspm_pm10, "RSPM/PM10"]



print(f"Location with the highest average SO2 level: {highest_so2}")
print(f"Mean SO2 level at this location: {mean_so2_highest}")

```

```python
print(f"Location with the highest average NO2 level: {highest_no2}")
print(f"Mean NO2 level at this location: {mean_no2_highest}")
```

```python

print(f"Location with the highest average RSPM/PM10 level: {highest_rspm_pm10}")
print(f"Mean RSPM/PM10 level at this location: {mean_rspm_pm10_highest}")
```

```python
#ii) Create visualizations :Create visualizations using data visualization libraries (e.g., Matplotlib, Seaborn).

#Histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['SO2'], bins=20)
plt.title('Histogram of SO2')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
#Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(df['City/Town/Village/Area'], df['RSPM/PM10'], alpha=0.5, color='green')
plt.title('Scatter Plot ')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#line Plot
plt.figure(figsize=(8, 6))
sns.lineplot(x='Sampling Date', y='NO2', data=df)
plt.xlabel('Category')
plt.ylabel('Numeric Value')
plt.show()
```
